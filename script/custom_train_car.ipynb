```python
import os
import cv2
import glob
import yaml
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from ultralytics import YOLO

import warnings
warnings.filterwarnings('ignore')
np.random.seed(12345)
```

```python
dataDir = '/home/ncrl/ncrl/yolo_ws/src/yolov8/CarDataset_colab/'
workingDir = '/home/ncrl/ncrl/yolo_ws/src/yolov8/CarDataset_colab/'
```

```python
for mode in ['train', 'valid']:
  print(f' \nImage sizes in {mode} set:\n')
  image_size = 0
  for file in glob.glob(os.path.join(dataDir, 'images', mode, '*')):
    image = Image.open(file)
    if image.size != image_size:
      print(image.size)
      image_size = image.size
```

```python
for mode in ['train', 'valid']:
  files = glob.glob(os.path.join(dataDir, 'images', mode,'*'))
  print(f'{mode} set size: {len(files)}\n')
```

### Downwloading the YOLOv8 Model

```python
model = YOLO('yolov8n.pt')
```

### Train the Model

```python
model.train(data=os.path.join(workingDir, 'data.yaml'),
            task = 'detect',
            imgsz = 640,
            epochs = 100,
            batch = 16,
            mode = 'train',
            name = 'yolov8_trained_model')
```
